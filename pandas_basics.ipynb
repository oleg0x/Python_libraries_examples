```python
import numpy as np
import pandas as pd
```

```python
# class pandas.Series(data=None, index=None, dtype=None, name=None, copy=None, fastpath=<no_default>)
# 1-dim ndarray with axis labels (including time series).

s1 = pd.Series()                        # Create empty Series

s2 = pd.Series([1, 3, 5, np.nan, 8])    # Create Series from a list

dic = {'aaa': 1000, 'bbb': 2000, 'ccc': 3000}
s3 = pd.Series(data=dic)                # Create Series from a dictionary

s4 = pd.Series(np.random.rand(4), index=['a', 'b', 'c', 'd'], dtype=np.float32)

arr = np.array([1, 2, 3])
s5 = pd.Series(arr, copy=False)         # Create Series (view) from a 1-dim ndarray

print(s1, '\n\n', s2, '\n\n', s3, '\n\n', s4, '\n\n', s5)
```

```python
# class pandas.DataFrame(data=None, index=None, columns=None, dtype=None, copy=None)
# 2-dim, size-mutable, potentially heterogeneous tabular data.

df1 = pd.DataFrame()                    # Create empty DataFrame

dic = {'aaaa': [1, 2, 3], 'bbbb': [4, 5, 6], 'cccc': [7, 8, 9]}         # Create DataFrame from a dictionary
df2 = pd.DataFrame(data=dic, index=['AA', 'BB', 'CC'])

dic = {'aaa': [11, 21, 31, 41], 'bbb': pd.Series([12, 22, 32, 42])}     # Create DataFrame from a dictionary including Series
df3 = pd.DataFrame(data=dic)

print(df1, '\n\n', df2, '\n\n', df3, '\n')
print(df2.bbbb)
```

```python
df = pd.DataFrame(
    {
        "AAA": 42,
        "BBB": pd.Timestamp("20250501"),
        "CCC": pd.Series(np.random.rand(5), index=list(range(100, 105)), dtype=np.float32),
        "DDD": np.array([3] * 5, dtype=np.int32),
        "EEE": pd.Categorical(["test", "train", "test", "train", "test"]),
        "FFF": "foobar"
    }
)

print(df, '\n')
print(df.dtypes)
```

```python
print(df.head(2), '\n')
print(df.tail(1))
```

```python
print(df.index)
print(df.columns)
```
