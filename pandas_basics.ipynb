```python
import numpy as np
import pandas as pd
```

```python
# class pandas.Series(data=None, index=None, dtype=None, name=None, copy=None, fastpath=<no_default>)
# 1-dim ndarray with axis labels (including time series).

s1 = pd.Series()                        # Create empty Series

s2 = pd.Series([1, 3, 5, np.nan, 8])    # Create Series from a list

dic = {'aaa': 1000, 'bbb': 2000, 'ccc': 3000}
s3 = pd.Series(data=dic)                # Create Series from a dictionary

s4 = pd.Series(np.random.rand(4), index=['a', 'b', 'c', 'd'], dtype=np.float32)

arr = np.array([1, 2, 3])
s5 = pd.Series(arr, copy=False)         # Create Series (view) from a 1-dim ndarray

print(s1, '\n\n', s2, '\n\n', s3, '\n\n', s4, '\n\n', s5)
```

```python
# class pandas.DataFrame(data=None, index=None, columns=None, dtype=None, copy=None)
# 2-dim, size-mutable, potentially heterogeneous tabular data.

df1 = pd.DataFrame()                    # Create empty DataFrame

dic = {'aaaa': [1, 2, 3], 'bbbb': [4, 5, 6], 'cccc': [7, 8, 9]}         # Create DataFrame from a dictionary
df2 = pd.DataFrame(data=dic, index=['AA', 'BB', 'CC'])

dic = {'aaa': [11, 21, 31, 41], 'bbb': pd.Series([12, 22, 32, 42])}     # Create DataFrame from a dictionary including Series
df3 = pd.DataFrame(data=dic)

print(df1, '\n\n', df2, '\n\n', df3, '\n')
print(df2.bbbb)
```

```python
dates = pd.date_range("20250501", periods=6)
print(dates)

df = pd.DataFrame(np.random.randn(6, 4), index=dates, columns=list("ABCD"))
print(df)
```

```python
print(df["B"], '\n')        # Select a column
print(df[1:4])              # Select some rows
```

```python
print(dates[3], '\n')

print(df.loc[dates[3]], '\n')           # Select a row matching a label

print(df.loc[dates[3], "B"])            # Select a single row and column label (returns a scalar)
print(df.at[dates[3], "B"], '\n')       # Select a single row and column label (returns a scalar)

print(df.loc[:, ["B", "D"]], '\n')      # Select all rows (:) with a select column labels
print(df.loc["20250502":"20250504", ["A", "C"]])
```

```python
print(df.iloc[3], '\n')             # Select a row matching a label
print(df.iloc[3:5, 1:3], '\n')      # Select specified rows and columns
print(df.iloc[[1, 3, 5], [0, 3]], '\n')

print(df.iloc[1:3, :], '\n')        # Slice rows explicitly
print(df.iloc[:, 1:3], '\n')        # Slice columns explicitly

print(df.iloc[4, 2])                # Get fast access to a scalar
print(df.iat[4, 2])                 # Get fast access to a scalar
```

```python
df = pd.DataFrame(np.random.randn(5, 3))
print(df, '\n')
pieces = [df[:3], df[3:]]           # Separate a DataFramerow-wise
print(pieces[0], '\n', pieces[1])
pd.concat(pieces)                   # Concatenate DataFrames together row-wise
print('\n', df)
```

```python
left = pd.DataFrame({"key": [101, 102, 103], "lval": [1, 1, 1]})
right = pd.DataFrame({"key": [101, 102, 103], "rval": [2, 2, 2]})
pd.merge(left, right, on="key")
```

```python
df = pd.DataFrame(
    {
        "A": ["foo", "bar", "foo", "bar", "foo", "bar", "foo", "foo"],
        "B": ["one", "one", "two", "three", "two", "two", "one", "three"],
        "C": np.random.randn(8),
        "D": np.random.randn(8),
        "E": np.random.randn(8),
    }
)

print(df, '\n')
print(df.groupby("B").sum(), '\n')                # Group by a column label and then apply sum() to the groups
print(df.groupby("A")[["C", "E"]].sum(), '\n')    # Group by a column label, select column labels, and then apply sum() to the groups
print(df.groupby(["A", "B"]).mean())         # Group by multiple columns label
```

```python
arrays = [
   ["bar", "bar", "baz", "baz", "foo", "foo", "qux", "qux"],
   ["one", "two", "one", "two", "one", "two", "one", "two"],
]

index = pd.MultiIndex.from_arrays(arrays, names=["First", "Second"])
df = pd.DataFrame(np.random.randn(8, 2), index=index, columns=["AAA", "BBB"])
print(df, '\n')

df2 = df[:4]
stacked = df2.stack(future_stack=True)      # “Compress” a level in the DataFrame’s columns
print(stacked)
```

```python
df = pd.DataFrame(
    {
        "A": ["one", "one", "two", "three"] * 3,
        "B": ["A", "B", "C"] * 4,
        "C": ["foo", "foo", "foo", "bar", "bar", "bar"] * 2,
        "D": np.random.randn(12),
        "E": np.random.randn(12),
    }
)
print(df, '\n')

t = pd.pivot_table(df, values="D", index=["A", "B"], columns=["C"])
print(t)
```

```python
df = pd.DataFrame(
    {"id": [1, 2, 3, 4, 5, 6], "raw_grade": ["a", "b", "b", "a", "a", "e"]}
)
print(df, '\n')

df["grade"] = df["raw_grade"].astype("category")
print(df["grade"], '\n')

new_categories = ["very good", "good", "very bad"]
df["grade"] = df["grade"].cat.rename_categories(new_categories)

df["grade"] = df["grade"].cat.set_categories(
    ["very bad", "bad", "medium", "good", "very good"]
)
print(df["grade"])
```
